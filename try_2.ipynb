```python
#open the file and skip rows
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('co2_weekly_mlo.csv',skiprows=47)
```

```python
df['roll'] = df['average'].rolling(12).mean()
print(df)
```

```python
from turtle import color
import matplotlib
from numpy import average
from matplotlib.pyplot import figure

figure(figsize=(18, 16), dpi=80)

ax1 =plt.subplot(211)
plt.title('ufh')
plt.xlabel('soa')
plt.plot(df['decimal'],df['average'],color='r')
plt.plot(df['decimal'],df['roll'],color='k')

ax2=plt.subplot(223)
plt.ylabel('Average Monthly CO2 (ppm)')
df['seasonal']=df['average']-df['roll']
df['seasonal'].groupby(df['day']).mean().plot(ax=ax2,kind='line')

plt.ylabel('Annual Increase in CO2 (ppm)')
df['seasonal'].groupby(df['year']).mean().plot(ax=ax2)
plt.show()
```
